```python
print(country_vaccinated_df[country_vaccinated_df['total_vaccinations']==country_vaccinated_df['total_vaccinations'].max()])
```

```python
# df1.groupby('Name')['Count'].max().reset_index()
largest_total_country = country_vaccinated_df.groupby('country')['total_vaccinations'].max().reset_index()
largest_total_country
```

```python
reset_index_largest_total = largest_total_country.nlargest(37, ['total_vaccinations']).reset_index(drop = True) 
reset_index_largest_total.index = reset_index_largest_total.index + 1
reset_index_largest_total
```

```python
new_country_df = reset_index_largest_total.country
new_total_vaccinations_df = reset_index_largest_total.total_vaccinations
```
